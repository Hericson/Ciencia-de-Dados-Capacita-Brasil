### 1. Introdução aos Princípios do Design de Visualizações

Visualizações são ferramentas poderosas para comunicar informações de forma clara e eficaz. Esta seção aborda os princípios de design de visualizações com exemplos de boas e más práticas, baseados em:

- Clareza

Além disso, destacaremos a importância de compreender o público-alvo e o contexto em que a visualização será utilizada.


### 1.1 Clareza

Visualizações claras permitem que os usuários entendam as informações rapidamente, sem ambiguidades. Evite sobrecarregar os gráficos com elementos desnecessários.


- Exemplo de má prática: Gráfico confuso

````python
# Dados para o gráfico
np.random.seed(10)
data = np.random.rand(10, 4)
categorias = ['A', 'B', 'C', 'D']

# Má prática: Gráfico confuso com excesso de elementos
plt.figure(figsize=(8, 6))
for i, col in enumerate(categorias):
    plt.plot(data[:, i], label=f"Categoria {col}")
plt.title("Exemplo de Gráfico Confuso")
plt.legend()
plt.show()
````

- Exemplo de má prática: Gráfico confuso

O primeiro exemplo demonstra um gráfico confuso devido a diversos problemas:

1. **Excesso de Linhas**: O uso de múltiplas linhas sobrepostas dificulta a identificação de padrões claros entre as categorias.
2. **Falta de Destaque**: Não há distinção suficiente entre os dados mais importantes e os elementos complementares.
3. **Legendas Exageradas**: O uso de uma legenda com muitas categorias sobrecarrega o leitor, tornando mais difícil identificar as linhas correspondentes.

Esses problemas podem levar à sobrecarga cognitiva, fazendo com que o público tenha dificuldade em interpretar as informações apresentadas.

---

- Exemplo de boa prática: Gráfico claro

````python

# Boa prática: Gráfico claro
plt.figure(figsize=(8, 6))
valoresMedios = data.mean(axis=0)
plt.bar(categorias, valoresMedios, color=['blue', 'orange', 'green', 'red'])
plt.title("Exemplo de Gráfico Claro")
plt.ylabel("Valores Médios")
plt.xlabel("Categorias")
plt.show()
````

- Exemplo de boa prática: Gráfico claro

No segundo exemplo, o gráfico é claro e eficaz pelos seguintes motivos:

1. **Simplicidade**: O gráfico apresenta apenas os valores médios das categorias, removendo informações redundantes ou desnecessárias.
2. **Uso de Barras**: As barras são mais adequadas para comparar valores entre categorias, tornando as diferenças mais visuais e óbvias.
3. **Cores Distintas**: As cores escolhidas ajudam a distinguir as categorias sem causar confusão ou distração.
4. **Elementos Relevantes**: Apenas os elementos essenciais, como título, eixos e barras, são exibidos, facilitando a compreensão do público.

Este gráfico prioriza a comunicação eficiente, garantindo que as informações sejam transmitidas de forma rápida e precisa.

### 1.2 Simplicidade
Gráficos simples são mais fáceis de interpretar. Remova elementos visuais desnecessários, como grades excessivas e legendas redundantes.


- Exemplo de má prática: Gráfico sobrecarregado

````python

# Dados para o gráfico
x = np.arange(1, 6)
y = [5, 7, 3, 8, 6]

# Má prática: Gráfico com excesso de informações
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Dados", marker='o', linestyle='--', color='purple')
plt.grid(which='both', color='red', linestyle='--', linewidth=1)
plt.legend(loc='upper right')
plt.title("Gráfico com Excesso de Informações")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
````

- Exemplo de má prática: Gráfico sobrecarregado

O primeiro exemplo apresenta um gráfico com excesso de elementos visuais desnecessários:

1. **Estilo Exagerado**: A combinação de linhas pontilhadas, marcadores e múltiplas cores adiciona complexidade sem contribuir para a compreensão dos dados.
2. **Excesso de Grades**: A presença de grades em todos os níveis visuais desvia a atenção dos dados principais.
3. **Legenda Supérflua**: Uma única série de dados foi rotulada na legenda, o que não é necessário neste caso, pois as informações podem ser deduzidas diretamente a partir do título e dos eixos.

Esses elementos extras poluem o design do gráfico, tornando-o visualmente pesado e dificultando a leitura para o público.


- Exemplo de boa prática: Gráfico simples

````python
# Boa prática: Gráfico simples
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', color='blue')
plt.title("Gráfico Simples")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
````

### 1.3 Uso de Cores
As cores devem ser usadas com propósito e moderação. Escolha paletas que sejam acessíveis para pessoas com daltonismo e evite combinações que confundam o leitor.

- Exemplo de má prática: Uso inadequado de cores

````python

# Má prática: Uso inadequado de cores
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [15, 25, 30, 10]

plt.figure(figsize=(8, 6))
plt.bar(categorias, valores, color=['red', 'green', 'blue', 'yellow'])
plt.title("Gráfico com Cores Confusas")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.show()
````

- O primeiro exemplo apresenta problemas significativos relacionados ao uso de cores:

1. **Combinações Distrativas**: O uso de cores como vermelho, verde, azul e amarelo cria uma combinação visualmente confusa e potencialmente problemática para pessoas com daltonismo.
2. **Falta de Propósito**: As cores não seguem uma lógica ou um esquema claro, tornando difícil entender se há alguma relação entre as categorias representadas.
3. **Estética Comprometida**: A escolha de cores saturadas e não harmoniosas dá ao gráfico uma aparência amadora, reduzindo sua eficácia em ambientes profissionais.

Essas escolhas comprometem a legibilidade e tornam a interpretação do gráfico mais difícil, especialmente para públicos diversos.


````python

# Boa prática: Uso apropriado de cores
plt.figure(figsize=(8, 6))
sns.barplot(x=categorias, y=valores, palette="viridis")
plt.title("Gráfico com Cores Apropriadas")
plt.ylabel("Valores")
plt.xlabel("Categorias")
plt.show()
````

No segundo exemplo, as cores são utilizadas de forma eficaz e intencional:

1. **Paleta Harmoniosa**: O uso de uma paleta consistente e visualmente agradável (como a "viridis") cria uma experiência visual mais fluida e acessível.
2. **Clareza e Contraste**: As cores escolhidas oferecem contraste suficiente para diferenciar as categorias sem causar distração.
3. **Inclusividade**: A paleta foi projetada para ser acessível a pessoas com daltonismo, garantindo que todos os públicos possam interpretar o gráfico adequadamente.

Este gráfico demonstra como o uso estratégico de cores pode melhorar a clareza, a estética e a acessibilidade, tornando a visualização mais eficiente e inclusiva para todos os públicos.

### 1.4 Público-Alvo e Contexto
É essencial compreender quem verá sua visualização e em que contexto ela será usada. Por exemplo:

- Um público técnico pode preferir gráficos mais detalhados.
- Um público geral pode precisar de gráficos mais simplificados.

Sempre ajuste seu design com base no objetivo da visualização.

- Público técnico: Scatter plot detalhado

````python

x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, size=x.shape)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.7, edgecolors='k')
plt.title("Visualização para Público Técnico")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

````

````python

x = np.linspace(0, 10, 100)
y_noise = np.sin(x) + np.random.normal(0, 0.1, size=x.shape)
y = np.sin(x)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

ax[0].scatter(x, y_noise, alpha=0.7, edgecolors='k', label='Noise', c='r')
ax[0].set_title('Sin + Noise')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].grid()

ax[1].scatter(x, y, alpha=0.7, edgecolors='k', label='Normal')
ax[1].set_title('Sin')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].grid()
plt.suptitle('Sin')




````

- Público geral: Linha suavizada

````python

plt.figure(figsize=(8, 6))
sns.lineplot(x=x, y=np.sin(x), color='blue', alpha=0.4)
plt.title("Visualização para Público Geral")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
````

### 2. DashBoards

### Conceito
Um ``dashboard`` é uma ferramenta visual que apresenta informações, dados e métricas de forma consolidada e interativa. 
Ele funciona como um "painel de controle", permitindo que os usuários monitorem indicadores-chave de desempenho (Key Performance Indicator- KPIs) e 
tomem decisões com base em dados em tempo real.

Dashboards são amplamente utilizados em diversas áreas devido à sua capacidade de traduzir dados complexos em visualizações intuitivas, como gráficos, tabelas e mapas. Isso facilita a interpretação de tendências, padrões e insights importantes.

---

### Por que Dashboards são importantes?
Dashboards são fundamentais para:
- ``Monitoramento em tempo real:`` Permitem visualizar o desempenho de processos, vendas ou outros indicadores à medida que os dados são gerados.
- ``Tomada de decisão:`` Apresentam informações claras e objetivas para suportar decisões estratégicas.
- ``Comunicação de dados:`` Facilitam o compartilhamento de resultados com equipes e stakeholders de forma visualmente atraente.
- ``Eficiência operacional:`` Centralizam informações relevantes em um único local, reduzindo o tempo gasto na busca e interpretação de dados.

---

### Exemplos Práticos de Uso
1. Empresas monitorando vendas:
   - Acompanhamento de vendas diárias, mensais e anuais.
   - Identificação de produtos ou regiões com melhor desempenho.
   - Comparação de receitas e custos para avaliar a lucratividade.

2. Análise financeira:
   - Controle de despesas e receitas em tempo real.
   - Identificação de tendências financeiras, como aumentos ou quedas em receitas.
   - Projeções de fluxo de caixa e lucros.

3. Controle de operações:
   - Monitoramento de estoques em tempo real.
   - Análise de tempo de entrega de pedidos.
   - Detecção de gargalos em processos operacionais.

````python
# Importando as bibliotecas necessárias
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# Criando o aplicativo Dash
app = Dash(__name__)

# Criando um dataset para exibição
df = px.data.gapminder()  # Dataset embutido no Plotly

# Layout do aplicativo
app.layout = html.Div([
    html.H1("Dashboard Interativo: Dados de Países", style={'textAlign': 'center'}),
    dcc.Dropdown(
        id='dropdown-continent',
        options=[
            {'label': continente, 'value': continente}
            for continente in df['continent'].unique()
        ],
        value='Asia',  # Valor padrão
        clearable=False,
        style={'width': '50%', 'margin': '0 auto'}
    ),
    dcc.Graph(id='scatter-plot'),
    html.Div(id='output-text', style={'textAlign': 'center', 'marginTop': 20})
])

# Callback para atualizar o gráfico e o texto
@app.callback(
    [Output('scatter-plot', 'figure'),
     Output('output-text', 'children')],
    [Input('dropdown-continent', 'value')]
)
def update_dashboard(continente_selecionado):
    df_filtrado = df[df['continent'] == continente_selecionado]
    figura = px.scatter(
        df_filtrado,
        x='gdpPercap',
        y='lifeExp',
        color='country',
        size='pop',
        hover_name='country',
        title=f"Relação entre PIB e Expectativa de Vida ({continente_selecionado})",
        labels={
            'gdpPercap': 'PIB per capita',
            'lifeExp': 'Expectativa de Vida',
            'pop': 'População',
            'country': 'País'
        }
    )
    texto = f"Total de países exibidos: {df_filtrado['country'].nunique()}"
    return figura, texto

# Executando o aplicativo
if __name__ == '__main__':
    app.run_server(debug=True)
````
