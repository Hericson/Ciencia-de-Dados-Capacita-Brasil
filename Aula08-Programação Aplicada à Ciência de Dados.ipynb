### 2.1.1 Quando Usar Gráficos de Barras

**Gráficos de barras** são uma das formas mais comuns e eficazes de visualizar dados categóricos. Eles são úteis em diversas situações, especialmente quando você deseja destacar diferenças, tendências ou comparações entre categorias.

---

##### Comparar Categorias
Use gráficos de barras para comparar valores entre diferentes categorias de dados. Exemplos incluem:
- Vendas por região.
- Número de alunos em diferentes cursos.
- Gastos em diferentes departamentos de uma empresa.

---

##### Exibir Dados Categóricos
Se os dados estão organizados em categorias distintas (e.g., tipos de produtos, meses do ano), os gráficos de barras permitem apresentar cada categoria de forma visualmente separada e clara.

---

##### Identificar Tendências
Gráficos de barras podem ser usados para revelar padrões ou tendências dentro de um conjunto de categorias, como:
- Crescimento ao longo de períodos de tempo.
- Aumento ou diminuição de preferências em pesquisas.

---

##### Comparar Partes em Relação ao Todo
Gráficos de barras também são úteis para mostrar proporções, como:
- A participação de mercado de cada concorrente.
- Distribuição de orçamentos entre setores.

---

##### Dados Fáceis de Comparar
Sempre que você quiser apresentar dados que o público possa comparar rapidamente, como:
- Desempenho de diferentes equipes em uma competição.
- Taxas de conclusão de projetos em um ano.

---

##### Quando os Dados São Simples
Gráficos de barras são mais eficazes quando os dados não são excessivamente complexos. Se você tem um conjunto pequeno ou moderado de categorias e valores, eles são a escolha ideal.

---

### 2.1.2 Evite gráficos de barras quando:
- Os dados têm muitos valores ou categorias (ficam sobrecarregados).
- Os dados são contínuos, como temperatura ou tempo (gráficos de linha são mais adequados).
- Deseja mostrar hierarquias complexas (gráficos de árvore ou dendrogramas podem ser melhores).

---

Usados corretamente, gráficos de barras tornam as informações mais acessíveis e ajudam na tomada de decisões baseada em dados.

### 2.2.1 Quando Usar Gráficos de Histograma

**Histogramas** são uma ferramenta poderosa para visualizar a distribuição de um conjunto de dados contínuos. Eles ajudam a compreender padrões, variações e tendências de dados em diferentes intervalos ou "bins". Abaixo estão as principais situações em que o uso de um histograma é recomendado.

---

#### Analisar Distribuições de Dados
Histogramas são ideais para mostrar como os dados estão distribuídos. Exemplos incluem:
- Distribuição de alturas em uma população.
- Frequência de notas em uma prova.
- Intervalos de tempo de resposta em um sistema.

---

#### Detectar Padrões ou Tendências
Eles ajudam a identificar padrões, como:
- Dados simétricos ou assimétricos.
- Distribuições normais, uniformes ou enviesadas.
- A presença de picos, múltiplos modos ou outliers.

---

#### Comparar Frequências em Intervalos
Use histogramas para comparar a frequência de dados dentro de intervalos específicos, como:
- Faixas etárias em uma pesquisa.
- Níveis de renda por faixa de valores.
- Intervalos de peso em um estudo de saúde.

---

#### Identificar Outliers
Histogramas podem revelar valores que se destacam da maioria dos dados, permitindo a análise de possíveis anomalias.

---

#### Avaliar a Variabilidade dos Dados
A largura dos intervalos e as alturas das barras permitem entender a dispersão e a concentração dos dados em certas regiões.

---

### Características Principais de um Histograma
- **Eixos:**
  - O eixo X (horizontal) representa os intervalos (ou "bins").
  - O eixo Y (vertical) mostra a frequência ou densidade de dados em cada intervalo.
- **Bins:** Dividem os dados em segmentos de intervalo para análise.
- **Sem Espaços entre Barras:** Diferentemente de gráficos de barras, histogramas não têm espaçamento entre as barras, pois os intervalos são contínuos.

---

### Detalhes sobre a escolha da quantidade de bins

### Impacto do Aumento dos Bins em um Gráfico de Histograma

O número de **bins** em um histograma tem um impacto direto na forma como os dados são visualizados e interpretados. Alterar a quantidade de bins pode mudar significativamente a percepção da distribuição dos dados. Abaixo estão os principais impactos do aumento dos bins:

---

### **Impactos do Aumento dos Bins**

### **Maior Detalhamento**
- Quando o número de bins aumenta, cada bin cobre um intervalo menor de dados.
- Isso permite visualizar com mais precisão pequenas variações na distribuição, destacando detalhes e padrões sutis.

### **Risco de Ruído**
- O aumento excessivo do número de bins pode fazer com que o histograma mostre muito "ruído". Pequenos desvios nos dados podem parecer mais importantes do que realmente são.
- A visualização pode se tornar confusa e difícil de interpretar, especialmente em dados com poucas observações.

### **Melhor Identificação de Outliers**
- Com mais bins, outliers podem se tornar mais evidentes, já que intervalos menores ajudam a isolar valores extremos.

### **Relação com o Tamanho do Dataset**
- Se o dataset for pequeno e o número de bins for muito alto, muitos bins podem acabar ficando vazios ou com poucos dados, dificultando a interpretação da distribuição.

---

### **Escolha do Número de Bins**
- Existem regras estatísticas e métodos para escolher o número ideal de bins, como:
  - **Regra de Sturges:** Apropriada para datasets menores
  - **Regra de Freedman-Diaconis:** Usa o intervalo interquartílico para definir o tamanho do bin.
  - **Teste Iterativo:** Ajustar manualmente o número de bins para encontrar o equilíbrio entre detalhes e clareza.
---


```python

# Dados de exemplo
data = np.random.normal(loc=50, scale=10, size=1000)  # 1000 valores com distribuição normal

# Aplicando a Regra de Sturges
n = len(data)  # Tamanho do conjunto de dados
bins = int(np.ceil(np.log2(n) + 1))  # Calculando o número de bins

# Plotando o histograma
plt.hist(data, bins=bins, edgecolor='black')
plt.title(f"Histograma com Regra de Sturges (Bins: {bins})")
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.show()
```

### Quando Usar a Regra de Sturges

- Conjuntos de dados pequenos ou moderados: É adequada para datasets com menos de algumas milhares de observações.
- Distribuição aproximadamente normal: A regra funciona melhor em distribuições simétricas.

### Limitações da Regra de Sturges

- Conjuntos de dados grandes: Pode subestimar o número de bins, tornando o histograma muito "grosseiro".
- Distribuições complexas: Não lida bem com dados que têm múltiplos picos ou outliers significativos.

Apesar de suas limitações, a Regra de Sturges é rápida, simples e eficaz para uma análise inicial de dados.

```python
# Dados de exemplo
data = np.random.normal(loc=50, scale=10, size=1000)  # 1000 valores com distribuição normal

# Cálculo do IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Tamanho do bin pela Regra de Freedman-Diaconis
n = len(data)
bin_width = 2 * IQR / np.sqrt(n)

# Número de bins
bins = int((data.max() - data.min()) / bin_width)

# Plotando o histograma
plt.hist(data, bins=bins, edgecolor='k')
plt.title(f"Histograma com Regra de Freedman-Diaconis (Bins: {bins})")
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.show()
```

### Vantagens da Regra de Freedman-Diaconis

Adapta-se à dispersão dos dados. Evita excesso de detalhes ou simplificações extremas, garantindo um histograma equilibrado. 

Essa abordagem é particularmente útil em conjuntos de dados com distribuições não uniformes ou com outliers significativos.

### Limitações da Regra de Freedman-Diaconis

A Regra de Freedman-Diaconis é uma abordagem eficaz e robusta para determinar o tamanho dos bins, especialmente em dados contínuos com distribuições simples e razoavelmente grandes. No entanto, ela deve ser usada com cautela em situações com:

- Poucos dados.
- Distribuições multimodais ou com outliers extremos.
- Necessidade de computação em tempo real para grandes volumes de dados.

Nesses casos, pode ser necessário experimentar outras regras ou ajustar manualmente os bins para obter a melhor visualização.

### 2.2.2 Evite histogramas quando:

- Os intervalos de dados não têm relevância (não faz sentido agrupar valores contínuos sem contexto).
- Você precisa comparar várias distribuições diretamente (nesse caso, gráficos de linhas podem ser mais úteis).

---

Usados adequadamente, histogramas permitem uma análise detalhada e clara da distribuição dos dados, auxiliando na tomada de decisões informadas e na detecção de padrões importantes.

### 3.3.1 Quando Usar Gráficos de Dispersão

**Gráficos de dispersão** (ou scatter plots) são ferramentas essenciais para analisar a relação entre duas variáveis. Eles ajudam a identificar padrões, correlações e tendências nos dados, especialmente em análises bivariadas.

---

#### Explorar Relações Entre Variáveis
Gráficos de dispersão são ideais para examinar a relação entre duas variáveis quantitativas. Exemplos incluem:
- Altura vs. peso.
- Idade vs. renda.
- Tempo de estudo vs. desempenho em provas.

---

#### Identificar Padrões ou Tendências
Eles ajudam a identificar padrões, como:
- Relações lineares (positivas ou negativas).
- Relações não lineares.
- Ausência de correlação.

---

#### Detectar Outliers
Gráficos de dispersão são eficazes para identificar pontos que se desviam significativamente do padrão geral dos dados.

---

#### Visualizar a Densidade dos Dados
- Em regiões com muitos pontos, os gráficos de dispersão mostram a densidade, permitindo identificar aglomerações.
- Quando necessário, combine com transparência (alpha) para melhorar a visualização.

---

#### Comparar Múltiplas Categorias
Use cores ou formas diferentes para representar categorias distintas dentro de um mesmo gráfico. Isso é útil para:
- Comparar subgrupos em um dataset.
- Destacar agrupamentos específicos de dados.

---

### 3.3.2 Características Principais de um Gráfico de Dispersão
- **Eixos:**
  - O eixo X (horizontal) representa uma variável.
  - O eixo Y (vertical) representa outra variável.
- **Pontos:** Cada ponto no gráfico representa uma observação com valores para as duas variáveis.
- **Interpretação Visual:** A inclinação, dispersão e agrupamento dos pontos oferecem insights sobre os dados.

---

### 3.3.3 Evite gráficos de dispersão quando:
- Uma ou ambas as variáveis são categóricas (gráficos de barras ou colunas são mais adequados).
- Os dados são muito dispersos ou possuem muitos pontos sobrepostos, dificultando a interpretação (use gráficos de densidade ou heatmaps como alternativa).
- Você precisa representar mais de duas dimensões (considere gráficos 3D ou scatter plots com cores, tamanhos e formas adicionais).

---

Gráficos de dispersão são ferramentas versáteis que fornecem insights valiosos sobre a relação entre variáveis, tornando-os indispensáveis em análises exploratórias de dados.

### 4.1 Definição de estilos (Seaborn)

`sns.set_theme(style="whitegrid")`:  Define o tema padrão de visualização do Seaborn, ajustando as propriedades dos gráficos gerados. Neste caso, o tema especificado é "whitegrid", que adiciona uma grade sobre um fundo branco.

#### Estilos disponíveis no Seaborn:

- "darkgrid": Fundo escuro com grades claras.
- "whitegrid": Fundo branco com grades claras (ideal para gráficos detalhados).
- "dark": Fundo escuro sem grades.
- "white": Fundo branco sem grades.
- "ticks": Fundo limpo com marcadores de eixo proeminentes.



```python
# Dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Todos os estilos
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# Aplicar os diferentes estilos
for style in seaborn_styles:
    sns.set_theme(style=style)
    plt.figure()
    plt.plot(x, y, label=f"Estilo: {style}")
    plt.legend()
    plt.title(f"Exemplo com o tema '{style}'")
    plt.show()
```

#### Personalizações adicionais:

 Você pode configurar a paleta de cores e o contexto visual:

```python

sns.set_theme(style="whitegrid", palette="deep", context="notebook")
```

- palette: Define a paleta de cores (e.g., "pastel", "deep", "muted", "bright").

- context: Ajusta o contexto de apresentação:
    - "paper": Para relatórios ou publicações.
    - "notebook": Para uso padrão em notebooks.
    - "talk": Para apresentações.
    - "poster": Para gráficos grandes e impactantes.

```python
# Dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Todos os estilos
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]
palettes = ["pastel" , "deep", "muted", "bright"]
contexts = ["paper", "notebook", "talk", "poster"]

# Aplicar os diferentes estilos
for style in seaborn_styles:
    for palette in palettes:
        for context in contexts:
            sns.set_theme(style=style, palette=palette, context=context)
            plt.figure()
            plt.plot(x, y, label=f"Estilo: {style}, Pallete: {palette}, Context:{context}")
            plt.legend()
            plt.title(f"Exemplo com o tema '{style} Pallete: {palette}, Context:{context}'")
            plt.show()
```

- Quartis (Limites Superior e Inferior da Caixa):
    - A parte inferior da caixa representa o primeiro quartil (Q1): 25% dos dados estão abaixo desse valor.
    - A parte superior da caixa representa o terceiro quartil (Q3): 75% dos dados estão abaixo desse valor.

- Intervalo Interquartil (IQR):

    - A altura da caixa é a diferença entre Q3 e Q1 (IQR = Q3 - Q1) e mostra a dispersão central dos dados.

- Bigodes (Extensões Verticais):

    - Os bigodes mostram a variação dos dados fora da caixa. Por padrão, eles se estendem até o menor ou maior valor dentro de 1,5 vezes o IQR. Valores fora desse intervalo são considerados outliers.

- Outliers (Pontos Individuais):

    - Pontos fora dos bigodes são outliers, indicando valores muito baixos ou altos em relação à maioria dos dados.

## 5. Técnicas de Normalização e Transformação de Dados

### 5.1 Normalização dos dados ``MinMaxScaler``

```python

# Normalização de Dados
from sklearn.preprocessing import MinMaxScaler

# Dados fictícios
vendas = np.random.randint(50, 500, size=10)
lucros = np.random.randint(10, 200, size=10)
dados = pd.DataFrame({'Vendas': vendas, 'Lucros': lucros})

# Normalizando os dados
scaler = MinMaxScaler()
dados_normalizados = pd.DataFrame(scaler.fit_transform(dados), columns=dados.columns)

# Comparação
print("Dados originais:")
print(dados)
print("Dados normalizados:")
print(dados_normalizados)
```

### Justificativa para realizar normalização dos dados

1. Comparação de Variáveis em Escalas Diferentes

    Os dados de vendas e lucros estão em escalas diferentes:

    Vendas variam entre 50 e 500.
    Lucros variam entre 10 e 200.

    Se essas variáveis forem usadas em análises ou modelos de machine learning (como regressão, clustering ou PCA), as diferenças de escala podem levar a:
- Maior peso para a variável com escala maior (neste caso, vendas) devido aos seus valores numéricos mais altos.
- Menor relevância da variável com escala menor (lucros), porque os valores numéricos são menores.

    Normalização ajusta todas as variáveis para a mesma escala (geralmente de 0 a 1), permitindo que ambas contribuam igualmente para as análises ou cálculos.


2. Requisito de Modelos de Machine Learning

    Muitos algoritmos de aprendizado de máquina dependem da magnitude dos valores (por exemplo, distância euclidiana em K-means ou gradientes em redes neurais).
    
    Sem normalização, as variáveis com escalas maiores podem distorcer os resultados ou dificultar a convergência dos modelos.


3. Redução de Sensibilidade a Unidades de Medida
    Sem normalização, as unidades de medida (ex.: reais para vendas, porcentagem para lucros) podem influenciar desproporcionalmente a análise.

    Normalizar elimina a influência da unidade de medida e coloca todas as variáveis em uma escala comum.

### 5.2 Tranformando dados com ``replace``

```python
import pandas as pd

# Criando um DataFrame de exemplo
dados = pd.DataFrame({
    'Nome': ['Ana', 'João', 'Maria', 'Pedro', 'Luiza'],
    'Cidade': ['Fortaleza', 'São Paulo', 'Rio de Janeiro', 'Fortaleza', 'São Paulo'],
    'Salario': [2500, 3000, 4000, 2000, 5000]
})

# Exibindo o DataFrame original
print("DataFrame Original:")
print(dados)

# 1. Substituir Nomes de Cidades por Regiões
cidade_para_regiao = {
    'Fortaleza': 'Nordeste',
    'São Paulo': 'Sudeste',
    'Rio de Janeiro': 'Sudeste'
}
dados['Regiao'] = dados['Cidade'].replace(cidade_para_regiao)

# 2. Transformar Salários em Categorias com replace
# Definindo categorias manuais
dados['Categoria_Salario'] = dados['Salario'].replace({
    2000: 'Baixo',
    2500: 'Médio',
    3000: 'Médio',
    4000: 'Alto',
    5000: 'Muito Alto'
})

# Exibindo o DataFrame transformado
print("\nDataFrame Transformado:")
print(dados)
```

### 5.3 Tranformando dados com ``cut`` e ``qcut``

```python
import pandas as pd

# Criando um DataFrame de exemplo
dados = pd.DataFrame({
    'Nome': ['Ana', 'João', 'Maria', 'Pedro', 'Luiza'],
    'Idade': [23, 35, 45, 29, 31],
    'Salario': [2500, 3000, 4000, 2000, 5000]
})

# Exibindo o DataFrame original
print("DataFrame Original:")
print(dados)

# 1. Transformar Idades em Categorias (Faixas Etárias)
# Definindo os limites das categorias
bins = [0, 25, 35, 50]  # Faixas: 0-25, 26-35, 36-50
labels = ['Jovem', 'Adulto', 'Meia-Idade']  # Nomes das categorias
dados['Faixa_Etaria'] = pd.cut(dados['Idade'], bins=bins, labels=labels)

# 2. Transformar Salário em Quartis
dados['Categoria_Salario'] = pd.qcut(dados['Salario'], q=4, labels=['Baixo', 'Médio', 'Alto', 'Muito Alto'])

# Exibindo o DataFrame transformado
print("\nDataFrame Transformado:")
print(dados)
```
