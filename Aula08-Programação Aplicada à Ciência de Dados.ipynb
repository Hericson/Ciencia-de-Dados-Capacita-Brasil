### 2.1.1 Quando Usar Grficos de Barras

**Grficos de barras** so uma das formas mais comuns e eficazes de visualizar dados categricos. Eles so teis em diversas situaes, especialmente quando voc deseja destacar diferenas, tendncias ou comparaes entre categorias.

##### Comparar Categorias
Use grficos de barras para comparar valores entre diferentes categorias de dados. Exemplos incluem:
- Vendas por regio.
- Nmero de alunos em diferentes cursos.
- Gastos em diferentes departamentos de uma empresa.

---

##### Exibir Dados Categricos
Se os dados esto organizados em categorias distintas (e.g., tipos de produtos, meses do ano), os grficos de barras permitem apresentar cada categoria de forma visualmente separada e clara.

---

##### Identificar Tendncias
Grficos de barras podem ser usados para revelar padres ou tendncias dentro de um conjunto de categorias, como:
- Crescimento ao longo de perodos de tempo.
- Aumento ou diminuio de preferncias em pesquisas.

---

##### Comparar Partes em Relao ao Todo
Grficos de barras tambm so teis para mostrar propores, como:
- A participao de mercado de cada concorrente.
- Distribuio de oramentos entre setores.

---

##### Dados Fceis de Comparar
Sempre que voc quiser apresentar dados que o pblico possa comparar rapidamente, como:
- Desempenho de diferentes equipes em uma competio.
- Taxas de concluso de projetos em um ano.

---

##### Quando os Dados So Simples
Grficos de barras so mais eficazes quando os dados no so excessivamente complexos. Se voc tem um conjunto pequeno ou moderado de categorias e valores, eles so a escolha ideal.

---

### 2.1.2 Evite grficos de barras quando:

### 2.2.1 Quando Usar Grficos de Histograma

**Histogramas** so uma ferramenta poderosa para visualizar a distribuio de um conjunto de dados contnuos. Eles ajudam a compreender padres, variaes e tendncias de dados em diferentes intervalos ou "bins". Abaixo esto as principais situaes em que o uso de um histograma  recomendado.

---

#### Analisar Distribuies de Dados
Histogramas so ideais para mostrar como os dados esto distribudos. Exemplos incluem:
- Distribuio de alturas em uma populao.
- Frequncia de notas em uma prova.
- Intervalos de tempo de resposta em um sistema.

---

#### Detectar Padres ou Tendncias
Eles ajudam a identificar padres, como:
- Dados simtricos ou assimtricos.
- Distribuies normais, uniformes ou enviesadas.
- A presena de picos, mltiplos modos ou outliers.

---

#### Comparar Frequncias em Intervalos
Use histogramas para comparar a frequncia de dados dentro de intervalos especficos, como:
- Faixas etrias em uma pesquisa.
- Nveis de renda por faixa de valores.
- Intervalos de peso em um estudo de sade.

---

#### Identificar Outliers
Histogramas podem revelar valores que se destacam da maioria dos dados, permitindo a anlise de possveis anomalias.

---

#### Avaliar a Variabilidade dos Dados
A largura dos intervalos e as alturas das barras permitem entender a disperso e a concentrao dos dados em certas regies.

---

### Caractersticas Principais de um Histograma
- **Eixos:**
  - O eixo X (horizontal) representa os intervalos (ou "bins").
  - O eixo Y (vertical) mostra a frequncia ou densidade de dados em cada intervalo.
- **Bins:** Dividem os dados em segmentos de intervalo para anlise.
- **Sem Espaos entre Barras:** Diferentemente de grficos de barras, histogramas no tm espaamento entre as barras, pois os intervalos so contnuos.

---

### Detalhes sobre a escolha da quantidade de bins

### Impacto do Aumento dos Bins em um Grfico de Histograma

O nmero de **bins** em um histograma tem um impacto direto na forma como os dados so visualizados e interpretados. Alterar a quantidade de bins pode mudar significativamente a percepo da distribuio dos dados. Abaixo esto os principais impactos do aumento dos bins:

---

### **Impactos do Aumento dos Bins**

### **Maior Detalhamento**
- Quando o nmero de bins aumenta, cada bin cobre um intervalo menor de dados.
- Isso permite visualizar com mais preciso pequenas variaes na distribuio, destacando detalhes e padres sutis.

### **Risco de Rudo**
- O aumento excessivo do nmero de bins pode fazer com que o histograma mostre muito "rudo". Pequenos desvios nos dados podem parecer mais importantes do que realmente so.
- A visualizao pode se tornar confusa e difcil de interpretar, especialmente em dados com poucas observaes.

### **Melhor Identificao de Outliers**
- Com mais bins, outliers podem se tornar mais evidentes, j que intervalos menores ajudam a isolar valores extremos.

### **Relao com o Tamanho do Dataset**
- Se o dataset for pequeno e o nmero de bins for muito alto, muitos bins podem acabar ficando vazios ou com poucos dados, dificultando a interpretao da distribuio.

---

### **Escolha do Nmero de Bins**
- Existem regras estatsticas e mtodos para escolher o nmero ideal de bins, como:
  - **Regra de Sturges:** Apropriada para datasets menores
  - **Regra de Freedman-Diaconis:** Usa o intervalo interquartlico para definir o tamanho do bin.
  - **Teste Iterativo:** Ajustar manualmente o nmero de bins para encontrar o equilbrio entre detalhes e clareza.
---


```python

# Dados de exemplo
data = np.random.normal(loc=50, scale=10, size=1000)  # 1000 valores com distribuio normal

# Aplicando a Regra de Sturges
n = len(data)  # Tamanho do conjunto de dados
bins = int(np.ceil(np.log2(n) + 1))  # Calculando o nmero de bins

# Plotando o histograma
plt.hist(data, bins=bins, edgecolor='black')
plt.title(f"Histograma com Regra de Sturges (Bins: {bins})")
plt.xlabel("Valores")
plt.ylabel("Frequncia")
plt.show()
```

### Quando Usar a Regra de Sturges

- Conjuntos de dados pequenos ou moderados:  adequada para datasets com menos de algumas milhares de observaes.
- Distribuio aproximadamente normal: A regra funciona melhor em distribuies simtricas.

### Limitaes da Regra de Sturges

- Conjuntos de dados grandes: Pode subestimar o nmero de bins, tornando o histograma muito "grosseiro".
- Distribuies complexas: No lida bem com dados que tm mltiplos picos ou outliers significativos.

Apesar de suas limitaes, a Regra de Sturges  rpida, simples e eficaz para uma anlise inicial de dados.

```python
# Dados de exemplo
data = np.random.normal(loc=50, scale=10, size=1000)  # 1000 valores com distribuio normal

# Clculo do IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Tamanho do bin pela Regra de Freedman-Diaconis
n = len(data)
bin_width = 2 * IQR / np.sqrt(n)

# Nmero de bins
bins = int((data.max() - data.min()) / bin_width)

# Plotando o histograma
plt.hist(data, bins=bins, edgecolor='k')
plt.title(f"Histograma com Regra de Freedman-Diaconis (Bins: {bins})")
plt.xlabel("Valores")
plt.ylabel("Frequncia")
plt.show()
```

### Vantagens da Regra de Freedman-Diaconis

Adapta-se  disperso dos dados. Evita excesso de detalhes ou simplificaes extremas, garantindo um histograma equilibrado. 

Essa abordagem  particularmente til em conjuntos de dados com distribuies no uniformes ou com outliers significativos.

### Limitaes da Regra de Freedman-Diaconis

A Regra de Freedman-Diaconis  uma abordagem eficaz e robusta para determinar o tamanho dos bins, especialmente em dados contnuos com distribuies simples e razoavelmente grandes. No entanto, ela deve ser usada com cautela em situaes com:

- Poucos dados.
- Distribuies multimodais ou com outliers extremos.
- Necessidade de computao em tempo real para grandes volumes de dados.

Nesses casos, pode ser necessrio experimentar outras regras ou ajustar manualmente os bins para obter a melhor visualizao.

### 3.3.1 Quando Usar Grficos de Disperso

**Grficos de disperso** (ou scatter plots) so ferramentas essenciais para analisar a relao entre duas variveis. Eles ajudam a identificar padres, correlaes e tendncias nos dados, especialmente em anlises bivariadas.

---

#### Identificar Padres ou Tendncias
Eles ajudam a identificar padres, como:
- Relaes lineares (positivas ou negativas).
- Relaes no lineares.
- Ausncia de correlao.

---

#### Detectar Outliers
Grficos de disperso so eficazes para identificar pontos que se desviam significativamente do padro geral dos dados.

---

#### Visualizar a Densidade dos Dados
- Em regies com muitos pontos, os grficos de disperso mostram a densidade, permitindo identificar aglomeraes.
- Quando necessrio, combine com transparncia (alpha) para melhorar a visualizao.

---

#### Comparar Mltiplas Categorias

- Comparar subgrupos em um dataset.

  - O eixo X (horizontal) representa uma varivel.

- **Interpretao Visual:** A inclinao, disperso e agrupamento dos pontos oferecem insights sobre os dados.

---

### 3.3.3 Evite grficos de disperso quando:
- Uma ou ambas as variveis so categricas (grficos de barras ou colunas so mais adequados).
- Os dados so muito dispersos ou possuem muitos pontos sobrepostos, dificultando a interpretao (use grficos de densidade ou heatmaps como alternativa).
- Voc precisa representar mais de duas dimenses (considere grficos 3D ou scatter plots com cores, tamanhos e formas adicionais).

---

Grficos de disperso so ferramentas versteis que fornecem insights valiosos sobre a relao entre variveis, tornando-os indispensveis em anlises exploratrias de dados.

### 4.1 Definio de estilos (Seaborn)

`sns.set_theme(style="whitegrid")`:  Define o tema padro de visualizao do Seaborn, ajustando as propriedades dos grficos gerados. Neste caso, o tema especificado  "whitegrid", que adiciona uma grade sobre um fundo branco.

#### Estilos disponveis no Seaborn:

- "darkgrid": Fundo escuro com grades claras.
- "whitegrid": Fundo branco com grades claras (ideal para grficos detalhados).
- "dark": Fundo escuro sem grades.
- "white": Fundo branco sem grades.
- "ticks": Fundo limpo com marcadores de eixo proeminentes.



```python
# Dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Todos os estilos
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# Aplicar os diferentes estilos
for style in seaborn_styles:
    sns.set_theme(style=style)
    plt.figure()
    plt.plot(x, y, label=f"Estilo: {style}")
    plt.legend()
    plt.title(f"Exemplo com o tema '{style}'")
    plt.show()
```

#### Personalizaes adicionais:

 Voc pode configurar a paleta de cores e o contexto visual:

```python

sns.set_theme(style="whitegrid", palette="deep", context="notebook")
```

- palette: Define a paleta de cores (e.g., "pastel", "deep", "muted", "bright").

- context: Ajusta o contexto de apresentao:
    - "paper": Para relatrios ou publicaes.
    - "notebook": Para uso padro em notebooks.
    - "talk": Para apresentaes.
    - "poster": Para grficos grandes e impactantes.

```python
# Dados de exemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Todos os estilos
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]
palettes = ["pastel" , "deep", "muted", "bright"]
contexts = ["paper", "notebook", "talk", "poster"]

# Aplicar os diferentes estilos
for style in seaborn_styles:
    for palette in palettes:
        for context in contexts:
            sns.set_theme(style=style, palette=palette, context=context)
            plt.figure()
            plt.plot(x, y, label=f"Estilo: {style}, Pallete: {palette}, Context:{context}")
            plt.legend()
            plt.title(f"Exemplo com o tema '{style} Pallete: {palette}, Context:{context}'")
            plt.show()
```

## 5. Tcnicas de Normalizao e Transformao de Dados

### 5.1 Normalizao dos dados ``MinMaxScaler``

```python

# Normalizao de Dados
from sklearn.preprocessing import MinMaxScaler

# Dados fictcios
vendas = np.random.randint(50, 500, size=10)
lucros = np.random.randint(10, 200, size=10)
dados = pd.DataFrame({'Vendas': vendas, 'Lucros': lucros})

# Normalizando os dados
scaler = MinMaxScaler()
dados_normalizados = pd.DataFrame(scaler.fit_transform(dados), columns=dados.columns)

# Comparao
print("Dados originais:")
print(dados)
print("Dados normalizados:")
print(dados_normalizados)
```

### Justificativa para realizar normalizao dos dados

1. Comparao de Variveis em Escalas Diferentes

    Os dados de vendas e lucros esto em escalas diferentes:

    Vendas variam entre 50 e 500.
    Lucros variam entre 10 e 200.

    Se essas variveis forem usadas em anlises ou modelos de machine learning (como regresso, clustering ou PCA), as diferenas de escala podem levar a:
- Maior peso para a varivel com escala maior (neste caso, vendas) devido aos seus valores numricos mais altos.
- Menor relevncia da varivel com escala menor (lucros), porque os valores numricos so menores.

    Normalizao ajusta todas as variveis para a mesma escala (geralmente de 0 a 1), permitindo que ambas contribuam igualmente para as anlises ou clculos.


2. Requisito de Modelos de Machine Learning

    Muitos algoritmos de aprendizado de mquina dependem da magnitude dos valores (por exemplo, distncia euclidiana em K-means ou gradientes em redes neurais).
    
    Sem normalizao, as variveis com escalas maiores podem distorcer os resultados ou dificultar a convergncia dos modelos.


3. Reduo de Sensibilidade a Unidades de Medida
    Sem normalizao, as unidades de medida (ex.: reais para vendas, porcentagem para lucros) podem influenciar desproporcionalmente a anlise.

    Normalizar elimina a influncia da unidade de medida e coloca todas as variveis em uma escala comum.

### 5.2 Tranformando dados com ``replace``

```python
import pandas as pd

# Criando um DataFrame de exemplo
dados = pd.DataFrame({
    'Nome': ['Ana', 'Joo', 'Maria', 'Pedro', 'Luiza'],
    'Cidade': ['Fortaleza', 'So Paulo', 'Rio de Janeiro', 'Fortaleza', 'So Paulo'],
    'Salario': [2500, 3000, 4000, 2000, 5000]
})

# Exibindo o DataFrame original
print("DataFrame Original:")
print(dados)

# 1. Substituir Nomes de Cidades por Regies
cidade_para_regiao = {
    'Fortaleza': 'Nordeste',
    'So Paulo': 'Sudeste',
    'Rio de Janeiro': 'Sudeste'
}
dados['Regiao'] = dados['Cidade'].replace(cidade_para_regiao)

# 2. Transformar Salrios em Categorias com replace
# Definindo categorias manuais
dados['Categoria_Salario'] = dados['Salario'].replace({
    2000: 'Baixo',
    2500: 'Mdio',
    3000: 'Mdio',
    4000: 'Alto',
    5000: 'Muito Alto'
})

# Exibindo o DataFrame transformado
print("\nDataFrame Transformado:")
print(dados)
```

### 5.3 Tranformando dados com ``cut`` e ``qcut``

```python
import pandas as pd

# Criando um DataFrame de exemplo
dados = pd.DataFrame({
    'Nome': ['Ana', 'Joo', 'Maria', 'Pedro', 'Luiza'],
    'Idade': [23, 35, 45, 29, 31],
    'Salario': [2500, 3000, 4000, 2000, 5000]
})

# Exibindo o DataFrame original
print("DataFrame Original:")
print(dados)

# 1. Transformar Idades em Categorias (Faixas Etrias)
# Definindo os limites das categorias
bins = [0, 25, 35, 50]  # Faixas: 0-25, 26-35, 36-50
labels = ['Jovem', 'Adulto', 'Meia-Idade']  # Nomes das categorias
dados['Faixa_Etaria'] = pd.cut(dados['Idade'], bins=bins, labels=labels)

# 2. Transformar Salrio em Quartis
dados['Categoria_Salario'] = pd.qcut(dados['Salario'], q=4, labels=['Baixo', 'Mdio', 'Alto', 'Muito Alto'])

# Exibindo o DataFrame transformado
print("\nDataFrame Transformado:")
print(dados)
```
