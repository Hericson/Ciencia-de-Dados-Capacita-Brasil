```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('artificial1d.csv', header=None)
data.rename(columns={0: 'x', 1: 'y'}, inplace=True)
data
```

```python
X = data['x'].to_numpy().reshape(-1,1)
y = data['y'].to_numpy().reshape(-1,1)
N, D = X.shape
ones = np.ones(N).reshape(-1,1)

X = np.c_[ones, X]

x_test = np.linspace(-1,1, 1000).reshape(-1,1)
X_test = np.c_[np.ones(x_test.shape[0]), x_test]
X_test
```

```python
from logistic_regression import LinearRegression

lr = LinearRegression()

w = lr.fit(X,y)
y_predict = lr.predict(X_test, w)
```

```python
plt.scatter(data['x'], data['y'])
plt.grid(True)
plt.title('Gráfico de Dispersão')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x_test, y_predict, c='red')
```
